```python
def circle_area(radius):
    if radius < 0:
        raise ValueError(" Radius cann't be negative")
    pi = 3.14159
    area = pi * radius * radius
    return area
try:
    r = float(input("Radius of the Circle:"))
    print(f"The area of the cirle with radius {r} is {circle_area(r):.2f}")
except ValueError as e:
    print(f"Error {e}")
    
```

```python
def sum_numbers(*args):
    negatives = [num for num in args if num < 0]
    if negatives:
            raise ValueError(f"Negative numbers Detected: {negatives}")
    return sum(args)

try:
    print(f"Total: {sum_numbers(3,5,2,9)}")
except ValueError as e:
    print(f"Error {e}")
 
```
