FUNCTION with LIST

```python
def check_even_list(num_list):
    for number in num_list:
        if number % 2 == 0:
            return True
        else:
            pass
    return False

```

```python
check_even_list([1,3,5,9,14])
```

```python
def check_even_list_one(num_list):
    #return all the even number in a list
    even_number = []
    for number in num_list:
        if number % 2 == 0:
            even_number.append(number)
        else:
            pass

    return even_number
```

```python
check_even_list_one([1,3,5,7,8,9])

```

FUNCTIONS and TUPLE UNPACKING

```python
stock_price = [('APP:',400),('GOO:',500),('WIP:',560),('MIC:',780),('LAT:',680)]

for ticker,price in stock_price:
    print(ticker,price+(0.1*price))
```

```python
work_hours = [('Younus',500),('Sravan',560),('Saikumar',800),('Giridhar',679),('Jaya',980)]
def employee_check(work_hours):
    current_max = 0
    employee_of_month = ''

    for employee,hours in work_hours:
        if hours > current_max:
            current_max = hours
            employee_of_month = employee
        else:
            pass
    return(employee_of_month,current_max)
        
```

```python
employee_check(work_hours)
```

```python
student_marks = [('Zahid',50),('Zakir',50),('Younus',86),('Nisar',75),('Nayeem',55)]

def check_marks(student_marks):
    current_marks = 0
    first_class = ' ' 
    for student,marks in student_marks:
        if marks > current_marks:
            current_marks = marks
            first_class = student
        else:
            pass
        
    return(first_class,current_marks)
```

```python
check_marks(student_marks)
```

# INTERACTIONS Between FUNCTIONS

Typically a python script contains several functions interacting with each other.


```python
from random import shuffle
#mylist = [20,56,34,54,23,11,22,90]

def shuffle_list(mylist):
    shuffle(mylist)
    return mylist
result = shuffle_list(mylist)

print(result)
```

```python
mylist1 = ['0',' ',' ']

def shuffle_list(mylist1):
    shuffle(mylist1)
    return mylist1
result = shuffle_list(mylist1)
print(result)
```

```python
def player_guess():
    guess = ''
    while guess not in ['0','1','2']:
        guess = input("Pick a Number: 0,1 or 2")
    return int(guess)  # Casting it to Integer


#You may be wondering why are these strings and why are you casting it as an integer at the end?
#Well, that's because input returns and always returns a string.So even if you type in the integer zero, one, or two
#it's going to return it as the form of a string,which is why we're checking the string zero, one, and two.
#But I'm returning it as an integer since I will use it to directly index off this list.

```

```python
player_guess()
```

```python
def check_guess(mylist1,guess):
    if mylist1[guess] == '0': 
        print("Correct!")
        print(mylist1)
    else:
        print("Wrong Guess")
        print(mylist1)
```

```python
# INITIAL LIST
mylist1 = [' ','0',' ']
# SHUFFLE LIST
mixed_list = shuffle_list(mylist1)
# USER GUESS
guess = player_guess()
# CHECK GUESS
check_guess(mixed_list,guess)
```

# *args and ** kwargs

*args return back TUPLES

**kwargs return DICTIONARY

```python
def myfunc(a,b,c,d,e):
    #Return 5% of the sum of a,b,c,d,e
    return sum((a,b,c,d,e)) * 0.5

myfunc(10,2,4,5,3,7)
```

```python
def myfunc(*args):
    return sum((args)) * 0.05
myfunc(3,5,6,7,1)  #arbitary number of arguments

```

```python
def myfunc(*args):
    for item in args:
        print(item)
```

```python

myfunc(10,4,2,20)
```

```python
def myfunc(**kwargs):
    print(kwargs)
    if 'fruit' in kwargs:
        print('My fruit of choice is {}'.format(kwargs['fruit']))
    else:
        print ('I did not find any fruit here')


myfunc(fruit='apple',veggie='lettuce')
```

```python
def myfunc(*args,**kwargs):
    print(args)
    print(kwargs)
    print('I would like to eat {} {}'.format(args[0],kwargs['food']))


myfunc(1,2,10,fruit='orange',food='Biryani',animal='dog')
```
