```python
rows = 5
#Upper part of the diamond
for i in range(1, rows + 1):
    print(" " *  (rows - i), end="")
    for j in range(1, 2 * i):
        if j == 1 or j == 2 * i - 1:
            print("*", end =" ")
        else:
            print(" ",end="")
    print()

#Lower part of the Diamond
for i in range(rows -1, 0 ,-1):
    print(" " *  (rows - i), end="")
    for j in range(1, 2 * i):
        if j == 1 or j == 2 * i - 1:
            print("*", end =" ")
        else:
            print(" ",end="")
    print()
```

While loops will Continue to execute a block of code while some condition remains TRUE.



```python
x = 0
while x < 5: 
    print(f'The value is {x}')
    x = x + 1
else:
    print('X IS not less than 5')




```

BREAK , CONTINUE , PASS

break : Break out of the current closest enclosing loop.

continue : Goes to the top of teh close enclosing loop.

pass: Does nothing at all.

```python
x = [1,2,3,4,5]
for l in x:
    pass
print('Will contine later')
```

```python
myname = 'MOHAMMAD YOUNUS NAWAZ'
for l in myname:
    if l=='m' or l=='M':
        continue  # Goes to the top of the close enclosing loop.
    print(l,end='')
```

```python
#BREAK

x = 0
while x < 5:
    if x == 2:
        break # Breaks out of the current closest enclosing loop
    print(x)
    x +=1
```

COMPREHENSION in PYTHON

A comprehension is a concise way to construct new sequences (like lists, sets, or dictionaries) by looping over iterables and optionally applying conditions.
It’s essentially a one-liner shortcut  for loops.


```python
mystring = 'hello'
mylist1 = []
for letter in mystring:
    mylist1.append(letter)
print(mylist1)


```

# List Comprehension
SYNTAX
[expression for item in iterable if condition]




```python

mylist = 'hello'
mylist = [letter for letter in mylist]
print(mylist)
```

Set Comprehension

Similar to list comprehension, but creates a set (unique values).



```python


unique_letters = {letter for letter in "banana"}
# Output: {'b', 'a', 'n'}
print(unique_letters)
```

Dictionary Comprehension

  Creates dictionaries in one line.

```python
squares_dict = {x: x**2 for x in range(5)}
# Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}

print(squares_dict)
```

Generator Expression
- Similar syntax, but uses () instead of [].
- Produces values lazily (one at a time).


```python
gen = (x**2 for x in range(5))
print(next(gen))  # Output: 0
print(next(gen))  # Output: 1
```

List Comprehension

SYNTAX

[ [PART1] [PART2] [PART3]]

Example:

nums = [1,2,3,4,5,6]

[n**2 for n in nums]


```python
result = [x**2 for x in range(10) if x % 2 == 0]
# Output: [0, 4, 16, 36, 64]

print(result)
```
