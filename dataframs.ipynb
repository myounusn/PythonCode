```python
def multiply_num(lst):
    total = 1
    for number in lst:
        total = total * number
    return total



multiply_num([1,2,3,4])
```

```python
import time
from tqdm.notebook import tqdm
def pbar_func():
    pbar = tqdm(range(0,50),desc="Processing...",bar_format='{desc}|{bar}|{percentage:.1f}')

    for i in pbar:    
        time.sleep(0.1)
        if i == 49:
            pbar.set_description("Done")
    pbar.close()
    
## New Change
```
