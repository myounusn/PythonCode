 # MAP and FILTER

 map() applies a function to every element in an iterable

 filter() keeps only the elements that satisfy the condition.


```python
def find_square(num):
    return num ** 2
    
```

```python
mynum=[1,2,3,4,5]
for item in map(find_square,mynum):   # One way is to iterate through the loop
    print(item)
```

```python
list(map(find_square,mynum))
```

```python
#USING THE map() and filter()


from functools import reduce

def multiply_list(numbers):
    # Filter out non-numeric values (just in case)
    filtered = filter(lambda x: isinstance(x, (int, float)), numbers)
    
    # Map each number to itself (identity mapping, but shows usage of map)
    mapped = map(lambda x: x, filtered)
    
    # Reduce to multiply all numbers together
    product = reduce(lambda a, b: a * b, mapped, 1)
    
    return product

# Example usage
print(multiply_list([2, 3, 4]))          # Output: 24
print(multiply_list([5, 10, "hello"]))   # Output: 50 (ignores "hello")
print(multiply_list([]))                 # Output: 1
```

```python
# "Find the sum of all the multiples of 3 or 5 below 1000."
def solve_problem_1():
    total_sum = 0
    for i in range(1, 1000):
        if i % 3 == 0 or i % 5 == 0:
            total_sum += i
    return total_sum

print(solve_problem_1())
```

```python
# Improvement 1 : Return values instead of just printing  -- using the sum() function

def check_lower_upper(s):
    lowercase = sum(1 for char in s if char.islower())
    uppercase = sum(1 for char in s if char.isupper())
    return{"original":s,"lowerCase":lowercase,"UpperCase":uppercase}


result=check_lower_upper("My Name is Mohammad Younus Nawaz")
print(result,end='')
```

```python
#Improvement 2 : More Pythonic with collection.Counter

#Counter  - it counts how many times each elements appears in an iterable ( like a list,string or generator)


from collections import Counter

def check_lower_upper(s):
    counts = Counter("lower" if c.islower() else "upper" if c.isupper() else "other" for c in s)
    return{"Original":s, "lower":counts['lower'],"upper":counts['upper'],"others":counts['other']}


check_lower_upper("My name is Mohammad Younus!")
```

```python
#This function will calculate the sum of each character in a given phrase or string 

from collections import Counter

def count_letters(s):
    # Create a Counter object from the string
    counts = Counter(s)
    
    # Return only alphabetic characters (ignore spaces, digits, punctuation)
    return {char: count for char, count in counts.items() if char.isalpha()}

# Example usage
result = count_letters("Mohammad")
print(result)
```

```python
import string
def is_pangram(s):
    # Convert to lowercase and make a set of alphabetic characters
    letters = sorted(set(c.lower() for c in s if c.isalpha()))
    #print(letters)
    
    # Sorted list of letters found
    #sorted_letters = sorted(letters)
    #print(sorted_letters)
    
    # Check if all 26 letters are present
    is_pangram = set(string.ascii_lowercase).issubset(letters)
    
    print("Letters found (in order):", letters)
    return is_pangram

# Example usage
sentence = "The quick brown fox jumps over the lazy dog"
print("Is pangram?", is_pangram(sentence))
```
